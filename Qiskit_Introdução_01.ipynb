```python
import numpy as np
from scipy import linalg as lng
import matplotlib.pyplot as plt
```

```python
#base computacional
v0 = np.array([ [1],[0] ]) #qubit0
v1 = np.array([ [0],[1] ]) #qubit1
print(v0)
print(v1)
```

```python
#estado arbitrario |psi> = a.|0> + b.|1>
# |a|^2 + |b|^2 = 1 definindo o estado do qubit
# a.a^* = |a|^2

Psi = np.sqrt(0.7)*v0 + 1j*np.sqrt(0.3)*v1 # j é a raiz de -1   j E C
print(Psi)
```

```python
Had = (1/np.sqrt(2))*np.array( [ [1,1], [1,-1] ]) # Hadamard, sobreposicao de estados
print(Had)
```

```python
Had@v0
```

```python
Had@Psi #aplicação de Hadamard no estado psi
```

```python
Had@Psi
```

```python
#dinamica do estado
psi11=list() #|a(t)| esta funcao prepara o esfera de bloch dinamizada
psi21=list() #|b(t)|
time=list()
for k in range(640):
    t = 0.01*k
    U = lng.expm(-1j*Had*t)
    psi = U@v0

    time.append(t)
    psi11.append( np.abs(psi[0,0]))
    psi21.append( np.abs(psi[1,0]))
```

```python
plt.xlabel("time")
plt.ylabel("$|\psi_{00}(t)|$")
plt.plot(time,psi11)
# mudanca do estado quantico do qubit mudado com o tempo
```
